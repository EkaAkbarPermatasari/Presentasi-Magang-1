an open-source project dedicated to provide a Python framework for processing seismological data. It provides parsers for common file formats and seismological signal processing routines which allow the manipulation of seismological time series

http://docs.obspy.org/

## Waveform data
GSE2/GSE1, miniSEED, SAC, SEG-Y, SEG-2, SH Q/ASCII,SEISAN, IRIS TSPAIR & SLIST, WAV
## Inventory data
Dataless SEED, XML-SEED, RESP
## Data request clients
ArcLink/WebDC, IRIS DHI/Fissures, SeisHub, IRIS web service, NERIES Web Service

# Metadata
consist of:
1. Station Information
- Lat./Lot./Depth/Component
2. Event Information
- Lat./Lon./Depth/Origin Time
3. Start and End Time of Waveforms
4. Sampling rate

Obspy Data Types: Overview (all in obspy.core)
- UTCDateTime
replacement/extension for the Python datetime object, stores a time stamp
- Stats
extension of the Python dict object, stores header information of waveforms
- Trace 
stores a single-channel, continuous piece of waveform data, consisting of waveform data and header information
- Stream
list-like container object, stores multiple traces (e.g. Z, N, E traces of one station)

# Stream methods
- merge(), attempts to merge Trace object with the same ID 
- rotate(), rotates two- or three-component Stream object
- select(), returns a new Stream with Traces matching the selection

# Trace methods 
- detrend(), removes a linear trend from the data
- filter(), filters the data
- normalize(), normalize the data to its absolute maximum
- remove_response(), deconvolves the instrument response
- resample(), resamples the data in the frequency domain
- taper(), tapers the data
- trim(), cuts the data to given start and end time
- decimate(), decumates the data by an integer factor
- differentiate(), differentiates the data with respect to time
- integrate(), integrates the data with respect to time
- trigger(), runs a triggering algorithm on the data
